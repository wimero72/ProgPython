La belleza de la sintaxis de Python
Una de las características más celebradas de Python es su sintaxis limpia y legible, un marcado contraste con otros lenguajes que pueden parecer abarrotados de símbolos y estructuras rígidas. En Python, no encontrarás una maraña de puntos y comas o llaves que dicten bloques de código. En su lugar, Python se basa en la indentación, utilizandoespacios o tabuladores, para definir la estructura del código. Esto no sólo hace que el código Python se parezca más al inglés sencillo, sino que también impone un estilo consistente y organizado que es más fácil de entender y mantener.

Echemos un vistazo a un simple mensaje de saludo:

1
print("Hello, world!")

Esta sola línea hace el trabajo maravillosamente. Pero para apreciar realmente el poder de la indentación, considere este ejemplo que demuestra una condición (que se discutirá en una lectura futura):

1234
if grade == "A":
    print("Excellent work!")
else:
    print("Keep studying and you'll improve!")

En este código, la indentación es crucial por varias razones:

Sentencias condicionales: Las líneas sangradas bajo if y else dentro del código definen las diferentes acciones que Python debe tomar dependiendo de las condiciones cumplidas. La sangría separa claramente estos diferentes caminos de ejecución.

Legibilidad: Incluso sin saber Python, probablemente puedas adivinar lo que hace este código debido a la sangría. Agrupa visualmente acciones relacionadas, facilitando el seguimiento de la lógica.

Ahora, contrastemos esto con el código que carece de sangría:

El intérprete de Python es el núcleo de Python. Cuando escribes código Python, es esencialmente un conjunto de instrucciones escritas en texto legible por humanos. Ventajas:

Interpretado vs. Compilado: Python es un lenguaje interpretado, lo que significa que se ejecuta línea a línea. Esto puede ser más lento que los lenguajes compilados (como C++ o Java) para ciertas tareas de rendimiento crítico.

Sentencias: Los comandos que orquestan la sinfonía de tu código
Imagina que diriges una orquesta. Utilizas gestos y señales específicas para dirigir a cada músico, creando una sinfonía armoniosa. En Python, estas señales son comparables a las instrucciones. Las líneas individuales de código sirven como instrucciones, dictando las acciones que su ordenador debe realizar.

Considere este ejemplo:

age = 25
if age >= 18:
    print("You are eligible to vote.")
else:

En este fragmento de código, encontramos varios tipos de sentencias trabajando en concierto:

Sentencia de asignación
La línea age = 25 asigna el valor numérico 25 a la variable age. Esto es como crear un personaje en una obra de teatro, estableciendo su estado inicial. Las variables actúan como contenedores para almacenar datos, permitiéndole referenciar y manipular esos datos a lo largo de su programa. En este caso, la variable age contiene la edad de la persona que estamos evaluando.

Sentencias condicionales y operadores de comparación
Las sentencias if age >= 18: y else: introducen puntos de decisión. El ordenador evalúa la condición (age >= 18) y ejecuta el bloque de código correspondiente en función de si la condición es verdadera o falsa. Así es como creamos rutas de bifurcación en nuestros programas, como en una historia de "elige tu propia aventura". Las sentencias condicionales nos permiten adaptar el comportamiento de nuestro código en función de circunstancias específicas. Por ejemplo, en nuestro ejemplo sólo imprimimos el mensaje "You are eligible to vote" si la variable age cumple el requisito de edad mínima para votar.

El núcleo de estas condiciones reside en los operadores de comparación. En nuestro ejemplo, >= (mayor o igual que) es un operador de comparación que comprueba la relación entre la edad y los 18 años. Otros operadores de comparación habituales son == (igual a), != (no igual a), > (mayor que) y < (menor que). El operador !=, en concreto, comprueba si dos valores no son iguales, como if user_input != "quit": para continuar un programa hasta que se escriba una palabra específica. Estos operadores son fundamentales porque producen un resultado booleano (o True o False), que es exactamente lo que necesitan las sentencias condicionales para decidir qué camino tomar.

Es crucial recordar que un signo igual simple (=) se utiliza para asignar (dar un valor a una variable, como age = 25), mientras que un signo igual doble (==) se utiliza para comparar (comprobar si dos valores son iguales).

Sentencias de impresión
Las líneas print("You are eligible to vote.") y print("You are not yet eligible to vote.") muestran mensajes al usuario. Son nuestra forma de comunicar los resultados de nuestros cálculos o de proporcionar información al usuario. Las sentencias print son esenciales para interactuar con los usuarios, depurar código y presentar información en un formato legible.

Expresiones: Los cálculos que sustentan la lógica de tu código
Mientras que las sentencias son las órdenes que dirigen el programa, las expresiones son los cálculos que se producen dentro de esas órdenes. Son las ecuaciones matemáticas u operaciones lógicas que producen valores.

Operaciones aritméticas básicas

Addition +

Subtraction -

Multiplication *

Division /

Operaciones aritméticas menos frecuentes
Floor Division // Realiza la división y descarta cualquier resto, dando como resultado un número entero (así 7 // 4 es 1, ya que 4 entra en 7 una sola vez de forma uniforme). También puede denominarse división entera.

Modulo % Devuelve el resto después de la división entera (así 7 % 4 es 3, ya que la división tiene un resto de 3).

Exponentiation ** Eleva un número a la potencia de otro número. Para escribir24, o dos a la cuarta potencia, se escribe 2 ** 4.

0.2 y 50 son expresiones sencillas que representan valores numéricos. Estos valores pueden utilizarse directamente en cálculos o almacenarse en variables para su uso posterior.

1 - discount es una expresión más compleja que calcula el porcentaje de descuento (por ejemplo, un descuento del 20% significa que el precio es el 80% del coste original). Esto demuestra cómo se pueden combinar expresiones utilizando operadores como la resta. Los paréntesis se utilizan para controlar el orden de las operaciones, garantizando que el descuento se resta de 1 antes de que tenga lugar la multiplicación.

price * (1 - discount) es la expresión final que determina el final_price multiplicando el price original por el descuento calculado. Esto demuestra la versatilidad de las expresiones, que nos permiten realizar operaciones aritméticas y manipular variables para obtener resultados significativos.

Las expresiones son los engranajes y las palancas que mueven el funcionamiento interno de sus programas. Permiten manipular datos, realizar comparaciones y encontrar nueva información. Sentencias de decisión: Controlar el flujo del código
En el núcleo de toda decisión hay una pregunta fundamental: ¿Esta condición es verdadera o falsa? En programación, las sentencias de decisión implementan directamente este concepto fundamental de verdadero/falso, actuando como mecanismo de control del flujo del programa. Evalúan meticulosamente una condición, que puede ir desde una simple comparación (por ejemplo, ¿la temperatura es superior a 30?) hasta una expresión compleja en la que intervienen múltiples variables, operadores y conectores lógicos. El resultado de esta evaluación dicta las acciones subsiguientes del programa, guiándolo por el camino adecuado, de forma parecida a como una aguja de ferrocarril dirige los trenes por vías diferentes en función de su destino.

La sentencia de decisión más básica y utilizada es la sentencia if. Imagine que está construyendo un programa para determinar si un cliente califica para el envío gratuito basado en el total de su pedido. Podría emplear una sentencia if como ésta

Python
12
if order_total >= 50:
  print("Congratulations! You qualify for free shipping.")

En este caso, la condición order_total >= 50 comprueba si el valor del pedido del cliente alcanza el umbral de envío gratuito. Si esta condición se evalúa como verdadera, indicando que el total del pedido es 50 o superior, el programa ejecuta el bloque de código sangrado, mostrando el mensaje de celebración. Por el contrario, si el total del pedido es inferior a 50, la condición es falsa y el mensaje permanece oculto.

Sentencias de selección: Ampliación del espectro de elección
Mientras que las sentencias de decisión como if proporcionan una elección binaria (verdadero o falso), las sentencias de selección como if-elif-else amplían esta lógica introduciendo múltiples condiciones a evaluar. Esto permite que el programa ejecute diferentes bloques de código en función de la primera condición que coincida. Piense en ello como un despachador de trenes que gestiona los cambios en una vía: el despachador evalúa el destino de cada tren secuencialmente y lo dirige en consecuencia. Cada tren (condición) se evalúa uno a uno y, una vez elegida la vía (acción) adecuada, se detiene la evaluación.

La sentencia if-else ejemplifica este mecanismo de elección ampliado. A partir de nuestro ejemplo anterior, podemos incorporar un bloque else para proporcionar información alternativa a los clientes que no alcanzan el umbral de envío gratuito:

1234
if order_total >= 50:
  print("Congratulations! You qualify for free shipping.")
else:
  print("Add more items to your cart to reach the free shipping threshold!")

Ahora, nuestro programa no sólo reconoce a los clientes que reúnen los requisitos para el envío gratuito, sino que también anima a los que se quedan cortos a aumentar el valor de su pedido, lo que puede aumentar las ventas y la satisfacción del cliente.

Enriquecimiento de la selección con cláusulas elif
En el ámbito de los escenarios del mundo real, las decisiones a menudo implican más de dos posibilidades. Aquí es donde entra en juego la cláusula elif, abreviatura de "else if" (si no). A diferencia de una simple sentencia if, una estructura if-elif-else permite definir múltiples condiciones y sus correspondientes acciones. Python evalúa estas condiciones secuencialmente, de arriba a abajo, y se detiene tan pronto como encuentra una condición que es verdadera. Esto significa que sólo se ejecutará un bloque de código, incluso si varias condiciones podrían ser teóricamente verdaderas. Este comportamiento secuencial garantiza claridad y eficacia en los procesos de toma de decisiones, ya que elimina la necesidad de comprobaciones redundantes.

Profundización: Condiciones anidadas y lógica booleana
Las sentencias de decisión y selección pueden mejorarse aún más incorporando condiciones anidadas y lógica booleana. Las condiciones anidadas implican incrustar sentencias if o if-else dentro de otra if o if-else, creando una estructura jerárquica de toma de decisiones. Esto es similar a un complejo árbol de decisiones en un sistema de diagnóstico médico, en el que cada rama representa un síntoma o el resultado de una prueba, y el programa analiza el árbol basándose en las respuestas del paciente y los datos médicos para llegar al diagnóstico más probable.

La lógica booleana, con operadores como y, o, y no, permite combinar varias condiciones y crear reglas de decisión más expresivas y complejas. Introducción a los bucles y a las sentencias condicionales
La programación a menudo requiere ejecutar el mismo bloque de código varias veces o tomar decisiones basadas en condiciones específicas dentro de un programa. Aquí es donde resultan esenciales los bucles y las sentencias condicionales, elementos fundamentales en cualquier lenguaje de programación, incluido Python. Estas construcciones permiten una ejecución de código eficiente y dinámica, crucial para cualquier aspirante a programador. Bucles: El poder de la repetición
Imagine escribir la misma instrucción repetidamente. Tedioso, ¿verdad? Los bucles ofrecen una forma de ejecutar instrucciones repetidamente, evitando la redundancia y fomentando la eficiencia. Piense en imprimir saludos para 100 personas. Sin un bucle, esto implicaría 100 instrucciones de impresión distintas Los bucles agilizan este proceso, ahorrando tiempo y esfuerzo. Python ofrece dos tipos principales de bucles: for y while, cada uno con sus propias ventajas y aplicaciones.

for bucle: Iterar sobre secuencias
Un bucle for itera sobre una secuencia, como una lista, tupla o cadena, ejecutando un bloque de código por cada elemento de la secuencia. Esto resulta especialmente útil cuando se trabaja con colecciones de datos o cuando se conoce de antemano el número de iteraciones. Por ejemplo, para imprimir los 10 primeros números naturales, un bucle for puede iterar eficazmente sobre los números del 1 al 10, ejecutando la función print() para cada número.

12
for i in range(1, 11):
  print(i)
Restablecer
En este ejemplo, range(1, 11) genera una secuencia de números del 1 al 10. La variable de bucle i toma cada valor de la secuencia, y el bloque de código dentro del bucle (en este caso, print(i)) se ejecuta para cada valor. Esto demuestra la concisión y eficiencia de los bucles en el manejo de tareas repetitivas.

Pero los bucles for no se limitan sólo a los números. Piense en procesar una lista de nombres, calcular el precio total de los artículos de un carro de la compra o incluso analizar una cadena de caracteres. El bucle for proporciona una forma estructurada de recorrer estas secuencias y realizar acciones en cada elemento.

while bucle: Ejecución basada en condiciones
Un bucle while, por el contrario, ejecuta un bloque de código mientras se cumpla una condición especificada. Esto resulta útil cuando se desconoce de antemano el número de iteraciones o cuando el bucle debe continuar hasta que se cumpla una condición específica. Por ejemplo, considere un programa que solicite a un usuario una entrada hasta que introduzca una respuesta válida. Un bucle while puede pedir repetidamente la entrada y comprobar su validez, terminando sólo cuando se proporciona una entrada válida.

1234567
valid_input = False
while not valid_input:
  user_input = int(input("Please enter a number greater than 0: "))
  if user_input > 0:
    valid_input = True
  else:
    print("Invalid input. Please try again.")
Este ejemplo muestra la flexibilidad de los bucles while para manejar situaciones en las que la condición de finalización no está predeterminada. El bucle continúa ejecutándose hasta que la variable valid_input se convierte en True, asegurándose de que el programa recibe la entrada deseada antes de continuar.

Este tipo de bucle es particularmente útil en escenarios como juegos interactivos, simulaciones o procesamiento de datos en tiempo real, donde el programa necesita reaccionar continuamente a las condiciones cambiantes y a las interacciones del usuario.

Ventajas del uso de bucles
Los bucles son vitales en programación por varias razones. Reducen la duplicación de código, eliminando la necesidad de escribir el mismo código repetidamente. Esto ahorra tiempo y esfuerzo y mejora la legibilidad y el mantenimiento del código. Imagínese intentar encontrar un error donde el mismo código se repite cientos de veces Los bucles facilitan enormemente la gestión y actualización del código.

Los bucles también permiten una iteración eficaz sobre estructuras de datos como listas y diccionarios, lo que facilita el acceso y la manipulación de los datos. En lugar de acceder a cada elemento individualmente, los bucles proporcionan una forma racionalizada de acceder y realizar operaciones en cada elemento.

Además, automatizan tareas repetitivas, liberando a los programadores para que puedan centrarse en aspectos complejos del desarrollo, como el diseño de algoritmos y la resolución de problemas. Al automatizar las tareas, los bucles permiten a los programadores concentrarse en el desarrollo creativo de software.

Sentencias condicionales: Introducción a la toma de decisiones
Las sentencias condicionales permiten a los programas ejecutar distintos bloques de código en función de si una condición es verdadera o falsa. Esto introduce la capacidad de tomar decisiones y permite a los programas responder dinámicamente a diversas situaciones y entradas del usuario. Piense en ello como si fuera un libro de aventuras en el que el camino a seguir depende de las decisiones que tome. La función range() genera una secuencia de números, y el bucle for itera sobre esta secuencia. Este bucle for continúa ejecutando la sentencia print e incrementa la variable i automáticamente. Estos bucles son especialmente adecuados para situaciones en las que se conoce el número de iteraciones. Existen varias formas de utilizar la función range():

range(stop): Genera una secuencia de números empezando por 0 (por defecto) hasta, pero sin incluir, el valor stop, incrementándose en 1 cada vez. Así, range(5) haría un bucle del 0 al 4.

range(start, stop): Esta forma le permite especificar el valor inicial de la secuencia. Genera números desde el valor inicial hasta el valor final, pero sin incluirlo, aumentando en 1 cada vez. Así, range (1, 11) haría un bucle del 1 al 10.

range(start, stop, step): Esta es la forma más flexible, ya que permite definir el incremento (o decremento) entre los números de la secuencia. Genera números empezando por el inicio, continuando hasta (pero sin incluir) el final, con cada número incrementado (o decrementado si el paso es negativo) por paso. Así, range(1, 10, 2) genera 1, 3, 5, 7 y 9. Una versión que utiliza pasos negativos es range(10, 5, -1), que generaría 10, 9, 8, 7 y 6.

Bucles anidados: Amplificación de la iteración

1. Errores de sangría
Python es conocido por sus estrictas reglas de sangría. A diferencia de muchos otros lenguajes que usan llaves u otros delimitadores para definir bloques de código, Python confía en la sangría para indicar la estructura de tu código. Una sangría mal colocada puede conducir a un comportamiento inesperado o a errores directos.

Imagina una simple sentencia if-else en la que el bloque else tiene una sangría incorrecta:

123456
x = 0

if x > 5:
    print("x is greater than 5")
  else:  # Incorrect indentation
    print("x is less than or equal to 5")
Restablecer

En este caso, el bloque else es tratado como una sentencia separada fuera de la estructura if-else, conduciendo a una salida incorrecta.

Solución: Preste siempre una atención meticulosa a la sangría. Asegúrese de que las sentencias dentro de bucles, condicionales, funciones y clases están correctamente sangradas. La mayoría de los editores de código proporcionan indicaciones visuales para ayudarle a mantener una sangría coherente.

2. Ámbito de las variables
En Python, las variables tienen un ámbito, que define la región de tu código donde la variable es accesible. Un error común es intentar acceder a una variable fuera de su ámbito.

Considera este ejemplo:

123456
flag = False
if flag:
    x = 10
    print(x) # No problem here

print(x) # No error if flag = True
Restablecer
Aquí, la variable x está definida dentro de la sentencia if. Intentar acceder a ella fuera de la sentencia if produce un error, si la bandera es False. Ejecute el código y observe un error, y luego cambie la bandera a True. Observa que un camino conduce a un error.

Solución: Sea consciente de donde define sus variables. Si necesitas que una variable sea accesible en todo tu código, defínela en el ámbito global. Sin embargo, el uso excesivo de variables globales puede llevar a un código más difícil de mantener, así que utilízalas con criterio.

3. Valores inmutables
 En Python, las cadenas son inmutables, lo que significa que su contenido no puede ser cambiado después de ser creadas. El código siguiente intenta cambiar el primer carácter de la cadena "Hola" de 'H' a 'h'. Sin embargo, esto resultará en un TypeError porque las cadenas son inmutables.

Las excepciones son eventos que interrumpen el flujo normal de tu código. No manejar las excepciones puede hacer que tu programa se bloquee. Estos eventos pueden ser provocados por varios factores, como entradas inválidas, intentos de acceder a archivos inexistentes, o incluso simples errores matemáticos como dividir por cero, como se muestra en el ejemplo:

Solución:  Excepciones como ZeroDivisionError serán tratadas con más detalle en un punto posterior del curso. Cuando existe una posibilidad o una excepción (como intentar abrir un archivo que no existe, o referenciar una variable que no existe), usarás bloques try-except para manejar las excepciones con gracia, como este ejemplo de código:

 Para imprimir "leche", puede hacer referencia al elemento por el índice:

Esta lista tiene una longitud de 5, con elementos en las posiciones de índice 0 a 4. Si quieres ver cuántos elementos hay en una lista, existe una función integrada para este propósito, len. Puedes imprimir la longitud de la lista usando len:

	print(len(grocery_list))

¿Y si quiere imprimir el último valor de la lista? En este ejemplo, sabes que el último elemento es 4, así que podrías imprimir el valor en el índice 4:

	print(grocery_list[4])

Sin embargo, las listas no siempre son tan sencillas y pueden crecer y decrecer continuamente. Una forma más segura de imprimir el último elemento es hacer referencia al elemento en len(nombre_lista) - 1. En este caso, la longitud de la lista es 5, por lo que el último elemento es 4.

	print(grocery_list[len(grocery_list) - 1])

Otra forma de hacer esto de una manera más legible es hacerlo en dos pasos. Primero, calcular el último elemento y, a continuación, mostrarlo:

	last_index = len(grocery_list) - 1

	print(grocery_list[last_index])

Rebanar: Acceso a una parte de la lista
A veces, sólo necesita una parte de la lista. Imagina que sólo quieres ver los dos primeros elementos de una lista. Python te permite extraer una "porción" de tu lista, como ésta:

El 0:2 significa que queremos los elementos desde el índice 0 (que es "leche") hasta el índice 2, pero sin incluirlo. El resultado serían los valores leche y hummus.

Comprensión de listas: Un enfoque conciso
Una comprensión de lista es una forma concisa y potente en Python de crear nuevas listas basadas en iterables existentes (como listas, tuplas, cadenas, rangos, etc.). Proporciona una alternativa más legible y a menudo más eficiente que el uso de los bucles for tradicionales para crear listas, especialmente cuando la lógica para crear la nueva lista es relativamente sencilla:

[expression for item in iterable]

Desglosemos cada parte:

[ ... ]: Los corchetes significan que el resultado será una nueva lista.

expression: Esta es la operación o transformación que se aplica a cada elemento del iterable. El resultado de esta expresión para cada elemento se convertirá en un elemento de la nueva lista.

for item in iterable: Esta parte especifica el iterable con el que se está trabajando y asigna cada elemento del iterable a la variable item durante cada iteración. Es similar al comienzo de un bucle for.

Errores comunes con las listas
A veces, accidentalmente puedes intentar hacer algo con una lista que cause un error. Aquí hay un par de errores comunes:

IndexError: Esto ocurre cuando intentas acceder a un índice que no existe en la lista. Por ejemplo, si tu lista tiene 5 elementos, y tratas de acceder a list[10], obtendrás un IndexError.

ValueError: Esto ocurre cuando intentas utilizar un método de lista con un valor que no tiene sentido. Por ejemplo, si intentas eliminar() un elemento que no está en la lista, obtendrás un ValueError.

Listas bidimensionales: Listas dentro de listas
Al igual que puedes poner diferentes tipos de cosas en una lista, ¡puedes incluso poner otras listas dentro de una lista! Esto se llama "lista bidimensional" o "lista de listas" Es como tener un armario con cajones, y que cada cajón contenga distintos tipos de objetos. En un ejemplo del mundo real, si estuvieras creando un juego de damas, necesitarías utilizar una lista bidimensional.

Por ejemplo, imagina que quieres representar una cuadrícula de números:

	grid = [

	  [1, 2, 3],

	  [4, 5, 6],

	  [7, 8, 9]

	]

Operaciones avanzadas con listas
Python proporciona un rico conjunto de herramientas para trabajar con listas. Estas son algunas de las más comunes. Dada la lista de antes, he aquí algunos ejemplos:

	grocery_list = ["milk", "hummus", "bread", "cheese", "apples"]

len(lista): Devuelve el número de elementos de la lista. Por ejemplo, len(lista_comestibles) sería 5.

list.append(x): Añade x al final de la lista. Por ejemplo, lista_comestibles.append("plátanos") añadiría "plátanos" al final de la lista.

list.insert(i, x): Inserta x en el índice i de la lista. Por ejemplo, grocery_list.insert(1, "yogurt") añadiría "yogurt" en el índice 1, desplazando "hummus" y los demás elementos a la derecha.

list.remove(x): Elimina la primera aparición de x de la lista. Por ejemplo, si accidentalmente añadimos "leche" dos veces, grocery_list.remove("leche") eliminaría la primera aparición.

list.sort(): Ordena los elementos de la lista en orden ascendente. Por ejemplo, numbers.sort() ordenaría los números de menor a mayor, o las cadenas por orden alfabético. Sin embargo, ten en cuenta que sólo puedes ordenar listas que contengan todos los números o todas las cadenas, ¡no una mezcla de diferentes tipos!

list.reverse(): Invierte el orden de los elementos de la lista. Por ejemplo, si tenemos números = [1, 2, 3], entonces números.reverse() lo convertiría en [3, 2, 1].

list.count(x): Cuenta cuántas veces aparece x en la lista. Por ejemplo, si tenemos letras = ["a", "b", "a", "c", "a"], entonces letras.count("a") devolvería 3.
